```python
frames1 = pd.concat([df, df1,df2, df3,df4, df5, df6, df7, df8, df9],ignore_index=False)
print(frames1)
```

```python
frames2=frames1.drop(frames1.iloc[:, 18:26], inplace = True, axis = 1) 
```

```python
print(frames1)
```

```python
frames3=frames1.drop(['R','R.1','R.2','R.3','R.4'], axis = 1) 
```

```python
print(frames3)
```

```python
class_list=[]
for p in frames3[['pinky','ring','middle','palm','index','thumb']].values:
    c=0
    for i in range(0,5):
        if(p[i]!=0):
            c=c+1
    if(c>=2):
        for f_sum in p:
            f_sum+=f_sum
        if(f_sum>3):
            class_list.append('No Slip')
            
        else:
            class_list.append('Slip')
    else:
        class_list.append('Slip')
    
```

```python
frames3['Class']=class_list
```

```python
print(frames3)
```

```python
print(frames3.Class)
```

```python
X = frames3.drop('Class', axis=1)
y = frames3['Class']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.20)
```

```python
from sklearn.svm import SVC
svclassifier = SVC(kernel='poly')
svclassifier.fit(X_train, y_train)
```

```python
y_pred = svclassifier.predict(X_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
print(confusion_matrix(y_test,y_pred))
print(classification_report(y_test,y_pred))
```
